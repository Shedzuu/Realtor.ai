````python
from openai import OpenAI
import json

client = OpenAI(api_key="", base_url="https://api.deepseek.com")

def get_real_estate_details(client_prompt):
    response = client.chat.completions.create(
        model="deepseek-chat",
        messages=[
            {"role": "system", "content": "You are a professional realtor." +
             "Extract the following real estate features from the user's input and make sure values are in Russian and return them as a Python dictionary:\n\n" +
                "{ 'rooms': value, 'adress': value, 'price': value, 'city_region': value, 'floor': value, 'area': value, 'kitchen_studio': value," + 
                " 'apartment_condition': value, 'is_apartment_has_furniture(full or no)': value, 'is_previously_dormitory': value, 'security': value," + 
                " 'furniture_detailed': value, 'facilities': value, 'bathroom': value, 'zhiloi_complex': value, 'house_year': value," + 
                " 'parking': value, 'is_separated_toilet': value, 'toilet_count': value, 'description': value }\n\n" + 
                "If any information is missing, replace it with 'No Information'. Ensure the response is a properly formatted JSON string."},
            {"role": "user", "content": client_prompt},
        ],
        stream=False
    )
    return response.choices[0].message.content

# Example usage
client_input = "Ищу 2-комнатную квартиру в Алматы, Алмалинский район, желательно в районе Макатаева — Ади Шарипова. Интересует квартира с новым ремонтом и новой мебелью, кухня-студия. Общая площадь около 45 м², этаж не ниже 6-го. Важно чтобы квартира была полностью меблирована, сдавалась в долгосрочную аренду от 3 до 11 месяцев, я готов к оплате депозита. Не курю. Рассмотрю варианты с охраной, видеонаблюдением и домофоном. Бюджет до 290 000 тг + ком. услуги. Готов к заселению с 14 марта. Если есть подходящий вариант, пишите."
formatted_response = get_real_estate_details(client_input)
print(formatted_response)
````

````python
import json
import re

def extract_json_from_string(response_str):
    """
    Extracts and converts a JSON object from a formatted string containing markdown-style triple backticks.
    """
    # Remove markdown triple backticks and newlines
    json_match = re.search(r'```json\n(.*?)\n```', response_str, re.DOTALL)
    
    if json_match:
        json_str = json_match.group(1)
        try:
            return json.loads(json_str)
        except json.JSONDecodeError:
            return {"error": "Invalid JSON format"}
    else:
        return {"error": "No JSON found in response"}

real_estate_dict = extract_json_from_string(formatted_response)
print(real_estate_dict)
````

````python
import pandas as pd
import json
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def calculate_similarity_score(df, feature_dict, user_prompt):
    """
    Compares each row in the DataFrame with the feature dictionary using cosine similarity,
    ignoring features that contain 'No Information'. The 'description' feature is compared
    with the user prompt separately.
    """
    similarity_scores = []
    df = df.iloc[:, 1:]  # Ignore the first column (URL)
    
    # Filter out features that have "No Information" in the dictionary
    relevant_features = [col for col in df.columns if feature_dict.get(col, "No Information") != "No Information" and col != "description"]
    
    # Initialize vectorizer
    vectorizer = TfidfVectorizer()
    
    for index, row in df.iterrows():
        row_values = [str(row[col]) for col in relevant_features]
        dict_values = [str(feature_dict[col]) for col in relevant_features]
        
        # Compute TF-IDF vectors for feature comparison
        tfidf_matrix = vectorizer.fit_transform([" ".join(dict_values), " ".join(row_values)])
        feature_similarity = cosine_similarity(tfidf_matrix[0], tfidf_matrix[1])[0][0] if len(relevant_features) > 0 else 0
        
        # Compute TF-IDF similarity for 'description' using user_prompt
        description_similarity = 0
        if "description" in df.columns:
            description_text = str(row["description"])
            tfidf_desc_matrix = vectorizer.fit_transform([user_prompt, description_text])
            description_similarity = cosine_similarity(tfidf_desc_matrix[0], tfidf_desc_matrix[1])[0][0]
        
        # Sum feature similarity and description similarity
        total_similarity = feature_similarity + description_similarity
        similarity_scores.append(total_similarity)
    
    return similarity_scores

# Compute similarity scores
df["similarity_score"] = calculate_similarity_score(df, real_estate_dict, client_input)
````
